```python
def read_glove_vecs(glove_file):
    with open(glove_file, 'r', encoding="utf8") as f:
        words = set()
        word_to_vec_map = {}
        for line in f:
            line = line.strip().split()
            curr_word = line[0]
            words.add(curr_word)
            word_to_vec_map[curr_word] = np.array(line[1:], dtype=np.float64)
        
        i = 1
        words_to_index = {}
        index_to_words = {}
        for w in sorted(words):
            words_to_index[w] = i
            index_to_words[i] = w
            i = i + 1
    return words_to_index, index_to_words, word_to_vec_map
```

```python
word_to_ind, ind_to_word, word_to_emb = read_glove_vecs('glove.6B.50d.txt')
```

```python
words = np.genfromtxt('old_words.txt', delimiter = '\n', dtype = str)
emb_to_word = {}
def get_words_to_emb_data(words):
    data = []
    for w in words:
        data.append(word_to_emb[w])
    return np.array(data)

word_emb_data = get_words_to_emb_data(words)
```

```python
def emb_to_word(emb):
    for w in words:
        if list(emb) == list(word_to_emb[w]):
            return w
```

```python
def get_centroid_index(data, centroids):
    index_values = np.zeros((data.shape[0],), dtype = int)
    for i in range(data.shape[0]):
        min_distance = 1000000
        for c in range(centroids.shape[0]):
            distance = np.sum(np.square(data[i]-centroids[c], dtype = float))
            if distance < min_distance:
                min_distance = distance
                index_values[i] = c
    return index_values
```

```python
def compute_average(data, ind, K):
    new_centroid = np.zeros((K,data.shape[1]))
    for k in range(K):
        count = 0
        summ = np.zeros((data.shape[1],))
        for i in ind:
            if i == k:
                summ += data[i]
                count += 1
        if count != 0:
            new_centroid[k] = summ/count
        elif count == 0:
            new_centroid[k] = summ
        
    return new_centroid
```

```python
## Training and finding here
K = 50
centroids = np.random.randn(K, word_emb_data.shape[1])
for i in range(20):
    index_values = get_centroid_index(word_emb_data, centroids)
    centroids = compute_average(word_emb_data, index_values, K)
```
